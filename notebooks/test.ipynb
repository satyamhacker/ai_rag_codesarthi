```python
# Cell 1: Import necessary LangChain and local LLM modules
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.chat_models import ChatOllama
import time

# Cell 4: Initialize the LLM that will generate the final answer
llm = ChatOllama(model="mistral:7b", temperature=0)  # temperature=0 keeps answers factual

print("LLM Initialized.")

print("--- Task 1: Starting .invoke() ---")
start_time = time.time()

# Ek lamba prompt taaki AI ko sochne aur likhne mein thoda time lage
prompt = "how are you"

# 1. Action start: Yahan terminal ruk jayega
print("⏳ Waiting for the entire response... (Terminal is hanging)")

# ✅ Correct: Call the LLM to get a response
response = llm.invoke(prompt)

# response_metadata mein provider-specific extra details hoti hain
meta_info = response.response_metadata

print(meta_info)

# print("\n--- Backend Logs (Hidden from User) ---")
# print(f"Model Used: {meta_info.get('model_name')}")
# print(f"Finish Reason: {meta_info.get('finish_reason')}")
# print(f"Prompt Tokens: {meta_info.get('prompt_eval_count')}")
# print(f"Completion Tokens: {meta_info.get('eval_count')}")
# print(f"Total Duration: {meta_info.get('total_duration')} seconds")



print("\n\n✅ Done!")
```
