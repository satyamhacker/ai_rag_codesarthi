```python
from langchain_community.document_loaders.generic import GenericLoader
from langchain_community.document_loaders.parsers import LanguageParser
from langchain_text_splitters import Language

# Define the path to your main folder
folder_path = "../repo" # Replace this with your actual folder path

# Initialize the loader to search the whole directory
loader = GenericLoader.from_filesystem(
    path=folder_path,
    glob="**/*",          # The **/* tells it to look inside all sub-folders recursively
    suffixes=[".py"],     # Target only Python files
    parser=LanguageParser(language=Language.PYTHON, parser_threshold=500)
)

# Load all the code files into a list of documents
code_docs = loader.load()

print(f"Successfully loaded {len(code_docs)} Python documents from the folder.")

# Let's peek at the metadata of the first loaded file to see where it came from
if code_docs:
    print("\nFirst file source:", code_docs[0].metadata.get("source"))
```
