```python
from langchain_community.document_loaders import PyPDFLoader

# Initialize the loader with the file path
loader = PyPDFLoader("../pdf_file/mysql_cheatsheet.pdf")

# Load the PDF into a list of Document objects
pages = loader.load()

print(f"Loaded {len(pages)} pages.")
print("Content of page 1:", pages[0].page_content)
```

```python
from langchain_community.document_loaders import UnstructuredMarkdownLoader

# Initialize the loader
loader = UnstructuredMarkdownLoader("../repo/README.md")

# Load the markdown file
docs = loader.load()

print("Markdown content:", docs[0].page_content)
```
