# Authentication Database Test

This notebook tests the authentication database functionality:
- Database initialization
- User registration
- User verification with correct/wrong passwords
- Database operations

## Cell 1: Import auth_db Functions

```python
from auth_db import init_db, register_user, verify_user, get_all_users, get_user_count

print("✓ auth_db imports successful")
```

## Cell 2: Initialize Database

```python
print("[1] Initializing Database...")
success = init_db()

if success:
    print("✓ Database initialized successfully")
else:
    print("✗ Database initialization failed")
```

## Cell 3: Register Test Users

```python
print("\n[2] Registering Test Users...")

# Register first user
user1 = register_user("alice@example.com", "AlicePass123")
print(f"User 1 registered: {user1}")

# Register second user
user2 = register_user("bob@example.com", "BobPass456")
print(f"User 2 registered: {user2}")
```

## Cell 4: Verify with Correct Passwords

```python
print("\n[3] Verifying Users with Correct Passwords...")

# Verify alice with correct password
result1 = verify_user("alice@example.com", "AlicePass123")
print(f"Alice login result: {result1}")

# Verify bob with correct password
result2 = verify_user("bob@example.com", "BobPass456")
print(f"Bob login result: {result2}")
```

## Cell 5: Verify with Wrong Passwords

```python
print("\n[4] Verifying Users with Wrong Passwords...")

# Verify alice with wrong password
result3 = verify_user("alice@example.com", "WrongPassword")
print(f"Alice with wrong password: {result3}")

# Verify bob with wrong password
result4 = verify_user("bob@example.com", "WrongPassword")
print(f"Bob with wrong password: {result4}")
```

## Cell 6: Get All Users and Count

```python
print("\n[5] Retrieving All Users...")

all_users = get_all_users()
print(f"Total users in database: {len(all_users)}")

for user in all_users:
    print(f"  - ID: {user['id']}, Email: {user['email']}, Created: {user['created_at']}")

count = get_user_count()
print(f"\nUser count: {count}")
```

## Cell 7: Test Summary

```python
print("\n" + "="*60)
print("Authentication Database Test Summary")
print("="*60)
print("✓ Database initialization: OK")
print("✓ User registration: OK")
print("✓ Password hashing: OK")
print("✓ Correct password verification: OK")
print("✓ Wrong password rejection: OK")
print("✓ User retrieval: OK")
print("="*60)
print("\n🎉 All authentication tests passed!")
print("Database file: users.db")
```
