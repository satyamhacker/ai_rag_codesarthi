## Cell 1: Import LangChain and Ollama

```python
from langchain_ollama import ChatOllama
from langchain_community.llms import Ollama
import sys

print("✓ LangChain imports successful")
print(f"Python version: {sys.version}")
```

## Cell 2: Initialize Mistral 7B Model

```python
# Initialize the ChatOllama model
llm = ChatOllama(model="mistral:7b")

print("✓ ChatOllama model initialized")
print(f"Model: mistral:7b")
```

## Cell 3: Test LLM Invocation

```python
# Test the model with a simple question
response = llm.invoke("What is a unit test?")

print("✓ LLM invocation successful")
print(f"\nResponse:\n{response.content}")
```
